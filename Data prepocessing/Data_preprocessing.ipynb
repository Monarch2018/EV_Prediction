```python
import pandas as pd
```

```python
train_data = pd.read_csv('./data/base_train.csv',index_col=0)
test_data = pd.read_csv('./data/base_test.csv',index_col=0)
```

### train data preprocessing

```python
del_cols = ['electric_vehicle_buy_reason_envir',
                'electric_vehicle_buy_reason_save_money',
                'electric_vehicle_buy_reason_new_tech_interest',
                'electric_vehicle_buy_reason_performance',
                'electric_vehicle_buy_reason_quietness',
                'electric_vehicle_buy_reason_other_recommended',
                'electric_vehicle_buy_again',
                'vehicle_type',
                'model_year',
                'ownership_status',
                'electric_vehicle_satisfied',
                'electric_vehicle_features_liked',
                'dataid',
                'car1',
                'quick_charge_port',
                'localhour',
                'delivery_date',
                'any_electric_vehicle',
                'status']

train_data.drop(del_cols,axis=1,inplace=True)
# group data into bins
bins = [0,3,5,7]
train_data['family_size'] = pd.cut(train_data['family_size'],bins)
object_cols = train_data.columns[train_data.dtypes=='object']
# one hot encoding
data_dummy = pd.get_dummies(train_data[object_cols])
data_dummy2 = pd.get_dummies(train_data['family_size'])
# combine data
data3 = train_data.drop(object_cols,axis=1)
data4 = data3.drop('family_size',axis=1)
data_final = pd.concat([data4,data_dummy,data_dummy2],axis=1)
# save data
data_final.to_csv('./data/data_final_train.csv')
```

### test data preprocessing

```python
test_data.drop(del_cols,axis=1,inplace=True)
# group data into bins
bins = [0,3,5,7]
test_data['family_size'] = pd.cut(test_data['family_size'],bins)
object_cols = test_data.columns[test_data.dtypes=='object']
# one hot encoding
data_dummy = pd.get_dummies(test_data[object_cols])
data_dummy2 = pd.get_dummies(test_data['family_size'])
# combine data
data3 = test_data.drop(object_cols,axis=1)
data4 = data3.drop('family_size',axis=1)
data_final = pd.concat([data4,data_dummy,data_dummy2],axis=1)
# save data
data_final.to_csv('./data/data_final_test.csv')
```

```python
train=pd.read_csv('./data/data_final_train.csv', index_col =0)
test=pd.read_csv('./data/data_final_test.csv', index_col =0)
```

```python
train['city_Humble'] = [0]*train.shape[0]
test['city_Plano'] = [0]*test.shape[0]
test['city_Dallas'] = [0]*test.shape[0]
test['city_Houston'] = [0]*test.shape[0]
test['city_Plano'] = [0]*test.shape[0]
test['city_Round Rock'] = [0]*test.shape[0]
test['city_Coppell'] = [0]*test.shape[0]
test['education_level_Some college/trade/vocational school'] = [0]*test.shape[0]
test['total_annual_income_$20,000 - $34,999'] = [0]*test.shape[0]
```

```python
test = test[list(train.columns)]
```
